```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler, Normalizer
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
```

```python
df_data = pd.read_csv('titanic_train.csv')
df_data
```

```python
df_data.isnull().sum()
```

```python
df_data = df_data.drop('Cabin', axis = 1)
df_data = df_data.drop('Name', axis = 1)
df_data = df_data.fillna(df_data['Age'].mean())
df_data = df_data.fillna(0)
df_data.isnull().sum()
```

```python
df_data = pd.get_dummies(df_data)
df_data.head()
```

```python
y_data = df_data.pop('Survived')
x_data = df_data
```

```python
logreg = LogisticRegression(fit_intercept=True)
logreg.fit(x_data, y_data)

logreg.predict(x_data) == y_data
```

### Accuray , Error

```python
accuracy = sum(logreg.predict(x_data) == y_data) / len(y_data)
accuracy
```

```python
error = 1 - accuracy
error
```

```python
from sklearn.naive_bayes import MultinomialNB, BernoulliNB
from sklearn.model_selection import cross_val_score
from sklearn.metrics import classification_report, accuracy_score
```

### Naive_Bayes

```python
clf_M = MultinomialNB()
clf_M.fit(x_data, y_data)

clf_B = BernoulliNB()
clf_B.fit(x_data, y_data)

scores_M = cross_val_score(clf_M, x_data, y_data, cv=5)
print("Multinomial Accuracy: %0.2f (+/- %0.2f)" % (scores_M.mean(), scores_M.std() * 2))

scores_B = cross_val_score(clf_B, x_data, y_data, cv=5)
print("Bernoulli Accuracy: %0.2f (+/- %0.2f)" % (scores_B.mean(), scores_B.std() * 2))
```

### Pipe_lr

```python
pipe_lr = Pipeline(steps=[('scl', StandardScaler()), ('clf', LogisticRegression())])
pipe_lr.fit(X = x_data, y = y_data)
pipe_rf = Pipeline(steps=[('scl', StandardScaler()), ('clf', RandomForestClassifier(n_estimators=1000,n_jobs=1, criterion='entropy'))])
pipe_rf.fit(X = x_data, y = y_data)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_data, pipe_lr.predict(x_data)))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_data, pipe_rf.predict(x_data)))
```

```python
pipe_lr.predict_proba(x_data)[:,1]
```

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score

scores = pipe_lr.predict_proba(x_data)[:, 1]
y_real = y_data
fpr_lr, tpr_lr, thresholds = roc_curve(y_real, scores)

roc_auc_lr = roc_auc_score(y_real, scores)
roc_auc_lr
```

```python
import matplotlib.pyplot as plt

plt.figure()
lw = 2
plt.plot(fpr_lr, tpr_lr, color='darkorange', lw=lw, label='ROC curve (area = %0.2f)' % roc_auc_lr)

plt.plot(fpr_lr, tpr_rf, color='blue', lw=lw, label='ROC curve of class {0} (area = {1:0.2f})'
        ''.format('rf', roc_auc_rf))

plt.plot([0, 1], [])
```
